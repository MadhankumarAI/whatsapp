```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")#nano yolov8 model

model.train(
    data="/home/user02/Documents/fences/electric_fences/data.yaml",
    epochs=50,
    imgsz=416,
    batch=4,
    device=0 
)
```

```python
import os
model = YOLO("/home/user02/Documents/fences/electric_fences/runs/detect/train4/weights/best.pt")
results = model("/home/user02/Documents/fences/electric_fences/test/images")
for r in results:
    r.show()#display window
    out_path = f"/home/user02/Documents/fences/electric_fences/preds/{os.path.basename(r.path)}"
    r.save(filename=out_path)


```

```python

metrics = model.val(data="/home/user02/Documents/fences/electric_fences/data.yaml",
                    split='test', 
                    device=0)  
print(metrics) 


```
